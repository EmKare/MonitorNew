```python
lot = 0
for routes in route_list:    
    route = client.directions(coordinates = [[car_pos[1], car_pos[0]], [route_list[lot][1][1],route_list[lot][1][0]]], profile = 'driving-car', format = 'geojson',)
    A_list = [tuple(reversed(coord)) for coord in route['features'][0]['geometry']['coordinates']]
    A_list.insert(0, tuple(car_pos))
    A_list.insert(len(A_list), (route_list[lot][1][0],route_list[lot][1][1]))    
    fl.PolyLine(locations = A_list, color = colours[lot]).add_to(m)
    plugins.AntPath(locations=A_list, reverse="True", dash_array=[20, 30], tooltip=f"from car_pos to {route_list[lot][0]}",).add_to(m)  
    print(f"car to {route_list[lot][0]} has {len(A_list)} points")
    
    route_distance, start, end = 0, 0, 1
    while end < len(A_list):
        route_distance += distance.distance(A_list[start], A_list[end]).km
        start += 1
        end += 1
    
    print(f"Roadway : {route_distance:.2f} kms")
    print(f"Straight: {distance.distance(car_pos, route_list[lot][1]).km:.2f} kms")
    
    lot += 1
```
